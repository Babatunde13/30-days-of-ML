```python
#importing neccessary packages
import pandas as pd
from sklearn.datasets import load_boston
import numpy as np
```

```python
#get the preferred data
data = load_boston().data
```

```python
#turn the data to a table
data = pd.DataFrame(data)
```

```python
#get the headers from the data
data.columns = load_boston().feature_names
```

```python
#add the target value
data['PRICE'] = load_boston().target
```

```python
#View the correlation table
data.corr()
```

```python
#This function gets the top5 correlation of the features of a dataset
def getCorr(dataset):
    top5 = []
    cols = dataset.corr().columns
    for i in cols:
        for j in dataset.corr()[i]:
            if j != 1:
                if len(top5) < 5:
                    top5.append(j)
                else:
                    if j > abs(min(top5)):
                        top5.remove(min(top5))
                        top5.append(j)
    return sorted(top5)

getCorr(data)
```
