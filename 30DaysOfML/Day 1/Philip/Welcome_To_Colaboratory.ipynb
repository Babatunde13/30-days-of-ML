```
import pandas as pd
from sklearn.datasets import load_boston
import numpy
```

```
print(load_boston().DESCR)
```

```
df = pd.DataFrame(load_boston().data)
df.columns = load_boston().feature_names
df["TARGET"] = load_boston().target
```

```
df.corr()
```

```
def bestest(feature_name):
  cor = df.corr()
  index,k = {}, {}
  for i in cor:
    index.update({cor[i][feature_name]:i})
  for i in sorted(index)[::-1][1:6]:
    k.update({i:index[i]})
  return k
print(bestest("TARGET"))
```
