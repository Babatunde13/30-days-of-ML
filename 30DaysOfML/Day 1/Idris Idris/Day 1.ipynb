```python
def best_of_features(data, target_column, num_of_features):
    '''
    The function produces the best features that correlates with the target feature provided the number of features needed.
    Params: 
    data: the data provided 
    target_column: the target column to be predicted for
    num_of_features: The number of best correlating fatures that we want.
    
    Output: List of values of the best correlating features.
    '''
    corr_data = data.corr()[target_column]
    listing = [abs(i) for i in corr_data if i != 1]
    listing.sort()
    listing[-num_of_features:]
    
    return [j for i in listing[-num_of_features:] for j in data.columns if data.corr()['Price'].abs()[j] == i]
```

```python
best_of_features(boston_data, 'Price', 5)
```
