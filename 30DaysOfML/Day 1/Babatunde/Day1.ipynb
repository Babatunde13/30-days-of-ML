```python
def best_features(data, target, numOfFeature):
    '''
    A function named best_features which computes the n best features in a dataframe using the spearman test.
    
    Parameter: 
                Data,  the dataframe
                target, the target variable/ column name
                numOfFeature: The numOfFeatures to be returned
                
    Returns: the n best featurs in a list.
    '''
    best_feature = []
    corr = list(data.corr()[target][:-1])
    corr1 = [abs(x) for x in corr]
    for cor in corr1:
        if len(best_feature) < numOfFeature:
              best_feature.append(max(corr1))
              corr1.remove(max(corr1))
    best = []  
    for x in best_feature:
        if x in corr:
            best.append(x)
        if x not in corr: 
            best.append(-x)
    best = [list(data.corr()[target]).index(x) for x in best]
    d = list(enumerate(data.columns))
    return [j[1] for x in best for j in d if j[0] == x ]
```

```python
best_features(df, 'Price', 5)
```
